```python
import pandas as pd
import seaborn as sb
from IPython import get_ipython
from IPython.display import display
```

```python
import pandas as pd
datos=pd.read_csv("celsius.csv")
datos.head(7)
```

```python
datos.info()
```

```python
import seaborn as sb
sb.scatterplot(x="celsius",y="Fahrenheit",data=datos)
```

```python
# x son caracteristicas y y son etiquetas
x=datos["celsius"]
y=datos["Fahrenheit"]
```

```python
x_modificada=x.values.reshape(-1,1)
y_modificada=y.values.reshape(-1,1)
```

```python
from sklearn.linear_model import LinearRegression
modelo=LinearRegression()
modelo.fit(x_modificada,y_modificada)
```

```python
celsius=int(input("ingrese la temperatura en celsius: "))
fahrenheit=modelo.predict([[celsius]])
print(str(celsius)+" C = "+str(fahrenheit)+" F")
```
